Creating Numpy Array from stratch


    zeros, ones, full, arange, linspace

```python
#Mục đích hàm zeros để tạo trước 1 mảng có số hàng, số cột cố định
#bấm tổ hợp Ctrl + Shift + Space để hiển thị gợi ý tham số đầu vào
#Thường kiểu dữ liệu numpy là float, để ép kiểu: dtype = int
np.zeros([2,3])
```

```python
#ép kiểu về int
np.zeros([2,3], dtype=int)
```

```python
np.ones([5,6])
```

```python
np.ones([2,2])*5
```

```python
#creat an array filled with a linear sequence
#starting at 0, ending at 20, stepping by 2
#this is similar to the built-in range() function

np.arange(0,10,2)
```

```python
#Dùng full để tạo mảng và điền vào số bạn mong muốn
np.full([3,5],9)
```

```python
#linspace dùng để chia khoảng giữa của 2 số
#ví dụ bắt đầu bằng 1 kết thúc bằng 2 chia ra 3 đoạn
np.linspace(1,2, 3)
```

random

```python
#Nhập vào 
np.random.random((4,4))
```

```python
#Sử dụng seed (hạt giống) để cố định random
np.random.seed(0)
np.random.random((4,4))
```

```python
#Điền vào các tham số sau cho normal
#mean: giá trị trung bình
#standerd deviation: độ lệch chuẩn
#size: 2x2

np.random.normal(2,3,(2,2))
```

```python
#tạo ma trận 10x10 với các số nguyên ngẫu nhiên trong khoảng từ 2 đến 9
x1 = np.random.randint(2,9,(10,10))
print(x1)
```

array concatenation & splitting

```python
#nối 2 mảng với nhau
x = np.array([1,2,3])
y = np.array([4,5,6])
print(x,y)
```

```python
#sử dụng concatenation để nối mảng
np.concatenate((x,y))
```

```python
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[7,8,9],[10,11,12]])
```

```python
np.concatenate((x,y)) #axit = 0
```

```python
np.concatenate((x,y), axis = 1)
```

hàm vstack (xếp chồng lên nhau)

```python
np.vstack((x,y))
```

hàm hstack nối cột

```python
np.hstack((x,y))
```
