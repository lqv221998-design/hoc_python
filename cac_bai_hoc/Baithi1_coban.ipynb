SU DUNG PHUONG THUC CO SAN

```python
names ="an, binh , chi, dung, giang"
```

```python
bo_dau_phay = names.replace(",", "")
list_names = bo_dau_phay.split()
print(list_names)
```

```python
list_names.append("hoa")
print(list_names)
```

```python
#viết hoa chữ cái đầu:
viet_hoa = []
for i in list_names:
    viet_hoa.append(i[0].upper() + i[1:])
print(viet_hoa)
```

```python
#Sử dụng title và list comprehension
#Nếu tên của bạn có nhiều chữ (ví dụ: "nguyen van an"), hàm này sẽ viết hoa chữ cái đầu của mỗi từ.
[i.title() for i in list_names]
```

```python
#Chỉ viết hoa chữ cái đầu của mỗi str
[i.capitalize() for i in list_names]
```

```python

while True:
    try:
        n = int(input("Nhập số dương n: "))
        if n < 0:
            print("hãy nhập số dương")
        else:
            if n % 2 == 0:
                print("số chẵn")
            else:
                print("số lẻ")
                break
    except ValueError:
        print("hãy nhập số nguyên")

```

```python
while True:
    try:
        n = int(input("Nhập số n: "))
        i = 1
        while i < n:
            if i % 2 != 0:
                print(i)
            i += 1
        break
    except ValueError:
        print("nhập số nguyên!")
    
```

```python
def tbc (list_so):
    result = sum(list_so)/len(list_so)
    return result
while True:
    try:
        day_so = input("nhập dãy số:")
        xu_ly = list(map(int,day_so.split()))
        print(tbc(xu_ly))
        break
    except ValueError:
        print("Hãy nhập dãy số cách nhau bởi dấu cách")
    
```
