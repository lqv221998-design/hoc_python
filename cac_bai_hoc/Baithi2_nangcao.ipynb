```python
while True:
    try:
        a = int(input("hy nhp s nguyn:"))
        result = 10/a
        print(result)
        break
    except ValueError:
        print("hy nhp s nguyn")
    except ZeroDivisionError:
        print("nhp s khc 0")
    finally:
        print(" thc hin xong") #finally lun thc thi d chng trnh b li
```
