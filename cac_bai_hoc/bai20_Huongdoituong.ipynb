```python
# 1. Định nghĩa "Khuôn" (Class)
class Room:

    
    def __init__(self, name, area):
        # Thuộc tính (Dữ liệu)
        self.name = name
        self.area = area

    # Phương thức (Hành động/Hàm bên trong class)
    def check_standard(self):
        if self.area > 10:
            return f"{self.name} đạt tiêu chuẩn."
        else:
            return f"{self.name} quá nhỏ."

# 2. Tạo các "Đối tượng" cụ thể (Objects) từ khuôn Room
r1 = Room("Phòng khách", 25)
r2 = Room("Phòng kho", 8)

# 3. Sử dụng
print(r1.check_standard())
print(r2.check_standard())
```
