```python
#sử dụng vòng lặp in ra dấu sao tạo thành hình vuông rỗng bên trong
so = int(input("Nhập số: "))
for i in range(so):
    for x in range(so):
        if i == 0 and i == 1 or i == x or x == 0 or x == so-1:
            print("* ", end="")
        else:
            print("  ", end="")
    print()
```

```python
# raise dùng để báo lỗi 
def kiem_tra_tuoi(tuoi):
    if tuoi < 18:
        raise ValueError("Tuổi phải nhỏ hơn 18")
    else:
        print(tuoi)
kiem_tra_tuoi(18)
```
