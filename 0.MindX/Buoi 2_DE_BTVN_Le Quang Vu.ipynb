Bài 2: Viết chương trình nhập vào 1 số bất kỳ nhỏ hơn 99, in ra màn hình bảng cửu chương của số đó

##### Cách 1

```python
# Yêu cầu người dùng nhập đến khi đúng
while True:
    try:
        num = int(input("Nhập số"))
        print(f"bảng cửu chương {num}:")
        bang_cuu_chuong = 0
        for i in range(1,11):
            bang_cuu_chuong = num * i
            print(f" {num} x {i} = {bang_cuu_chuong}")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```

##### Cách 2

```python
# Người dùng chỉ được nhập 3 lần và dùng list comprehension
Bo_dem = 0
while True:
    Bo_dem += 1
    if Bo_dem > 4:
        print("Bạn đã nhập quá số 3 lần")
        break
    try:
        num = int(input("Nhập số"))
        print(f"Bảng cửu chương {num} là:")
        bang_cuu_chuong = [f"{num} x {i} = {num*i}" for i in range(1,11)]
        print(*bang_cuu_chuong,sep="\n")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```

##### Cách 3

```python
# Tạo hàm lấy giá trị đầu vào giới hạn số lần nhập 
def nhap_du_lieu(gioi_han = 4):
    bo_dem = 0
    while bo_dem < gioi_han:
        bo_dem += 1
        try:
            nhap_so = int(input('Nhập một số:'))
            return nhap_so
        except ValueError:
            print("Bạn phải nhập số nguyên")
    print(f"bạn đã nhập quá số {gioi_han} lần")
    return None

# Tạo hàm tính toán:
def bang_cuu_chuong(nhap_du_lieu):
    return[f"{nhap_du_lieu} x {i} = {nhap_du_lieu * i}"for i in range(1,11)]

# --- Chương trình chính (Main) ---
ket_qua_nhap = nhap_du_lieu()
print(f"bảng cửu chương {ket_qua_nhap}:")
print(*bang_cuu_chuong(ket_qua_nhap),sep="\n") # Dùng * để in danh sách dọc xuống dòng (unpacking)


```

4. Viết chương trình nhập vào ngày tháng năm sinh đầy đủ của 1 người, tính con số vận mệnh của người đó (thần số học á :v)

```python
nam = input("Nhập năm: ")
thang = input("nhập tháng")
ngay = input("Nhập ngày")
print(f"{nam}/{thang}/{ngay})")
cong_nam = sum([int(x) for x in nam])
cong_thang = sum([int(y) for y in thang])
cong_ngay = sum([int(z) for z in ngay])
print(f"Tổng riêng lẻ: [{cong_nam},{cong_thang},{cong_ngay}]")
nam_rut_gon = sum([int(a) for a in str(cong_nam)])
thang_rut_gon = sum([int(b) for b in str(cong_thang)])
ngay_rut_gon = sum([int(c) for c in str(cong_ngay)])
print(f"Rút gọn tổng riêng lẻ: [{nam_rut_gon},{thang_rut_gon},{ngay_rut_gon}]")
cong_ket_qua = sum([nam_rut_gon,thang_rut_gon,ngay_rut_gon])
print(f"Cộng kết quả: {cong_ket_qua}")
ket_qua_cuoi = sum([int(d) for d in str(cong_ket_qua)])
print(f"Thần số học: {ket_qua_cuoi}")

```

Bai 6: Viết chương trình nhập vào một số, tính xem số đó có phải là số nguyên tố hay không

```python
def kt_so_nguyen_to(n):
    if n <= 1:
        return False 
    for i in range(2, n): #Lặp qua từng số từ 2 đến n
        if n % i == 0: #Kiểm tra xem n có chia hết cho số nào ngoài nó không
            return False  
    return True

n = int(input("nhap so tu nhien: "))

if kt_so_nguyen_to(n):
    print(f"{n} la so nguyen to")
else:
    print(f"{n} khong phai so nguyen to")
```

Bài 7: Viết chương trình nhập vào một số, tính giai thừa của số đó (giai thừa của 1 số n ký hiệu là n! là tích của 1 * 2 * 3 *... * n)


```python
def tinh_giai_thua(n):
    if n < 0:
        return "Không tồn tại giai thừa của số âm"
    elif n == 0 or n == 1:
        return 1
    else:
        giai_thua = 1
        for i in range(2, n + 1):
            giai_thua *= i
        return giai_thua
n = int(input("Nhập một số nguyên dương: "))
ket_qua = tinh_giai_thua(n)
print(f"Giai thừa của {n} là: {ket_qua}")
    
```

Bài tập 8: Xây dựng theo yêu cầu sau:
- Class: Vật liệu
- Attribute: Tên, Đơn giá, Đơn vị
- Method: Tính Giá (công thức: Đơn giá * Số lượng)
- Tạo Object tên "Bê Tông" với đơn giá = 1200000 và đơn vị m3. Tính giá của Bê tông nếu số lượng =100
- Note: Có đoạn code Xử lý lỗi khi người dùng nhập liệu sai


```python
class Vat_lieu:
    def __init__(self,ten,don_gia,don_vi):
        self.ten = ten
        self.don_gia = don_gia
        self.don_vi = don_vi
    
    def tinh_gia(self,so_luong):
        thanh_tien = self.don_gia * so_luong
        return thanh_tien
    
tinh_tien = Vat_lieu("Bê tông",120,"m3")
print(tinh_tien.tinh_gia(10))
```
