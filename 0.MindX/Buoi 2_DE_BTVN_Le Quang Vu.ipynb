Bài 2: Viết chương trình nhập vào 1 số bất kỳ nhỏ hơn 99, in ra màn hình bảng cửu chương của số đó

##### Cách 1

```python
# Yêu cầu người dùng nhập đến khi đúng
while True:
    try:
        num = int(input("Nhập số"))
        print(f"bảng cửu chương {num}:")
        bang_cuu_chuong = 0
        for i in range(1,11):
            bang_cuu_chuong = num * i
            print(f" {num} x {i} = {bang_cuu_chuong}")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```

##### Cách 2

```python
# Người dùng chỉ được nhập 3 lần và dùng list comprehension
Bo_dem = 0
while True:
    Bo_dem += 1
    if Bo_dem > 4:
        print("Bạn đã nhập quá số 3 lần")
        break
    try:
        num = int(input("Nhập số"))
        print(f"Bảng cửu chương {num} là:")
        bang_cuu_chuong = [f"{num} x {i} = {num*i}" for i in range(1,11)]
        print(*bang_cuu_chuong,sep="\n")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```

##### Cách 3

```python
# Tạo hàm lấy giá trị đầu vào giới hạn số lần nhập 
def nhap_du_lieu(gioi_han = 4):
    bo_dem = 0
    while bo_dem < gioi_han:
        bo_dem += 1
        try:
            nhap_so = int(input('Nhập một số:'))
            return nhap_so
        except ValueError:
            print("Bạn phải nhập số nguyên")
    print(f"bạn đã nhập quá số {gioi_han} lần")
    return None

# Tạo hàm tính toán:
def bang_cuu_chuong(nhap_du_lieu):
    return[f"{nhap_du_lieu} x {i} = {nhap_du_lieu * i}"for i in range(1,11)]

# --- Chương trình chính (Main) ---
ket_qua_nhap = nhap_du_lieu()
print(f"bảng cửu chương {ket_qua_nhap}:")
print(*bang_cuu_chuong(ket_qua_nhap),sep="\n") # Dùng * để in danh sách dọc xuống dòng (unpacking)


```
