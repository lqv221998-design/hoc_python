```python
# Tìm ra cầu thủ có số thẻ vàng nhiều nhất
df_player.loc[df_player['YellowCards',''].idxmax()]
```

```python
# Tìm ra Cầu thủ có số thẻ đỏ nhiều nhất
df_player.loc[df_player['RedCards'].idxmax()]
```

```python
# Tìm ra cầu thủ vừa có thẻ vàng nhiều nhất và ra trận nhiều nhất
df_player.loc[df_player['YellowCards'].max() & df_player['Appearances'].max()]
```
