#### Chương trình 1: Chương trình đơn giản, kèm xử lý ngoại lệ

```python
#Sử dụng hàm input, lệnh print, tính tổng, xử lý ngoại lệ:
a = input("Bạn hãy nhập giá trị đầu tiên: ") #Lỗi ValueError khi nhập chữ
b = input("Bạn hãy nhập giá trị thứ hai:")   #Lỗi ValueError khi nhập chữ
try:
    print(f"Tổng hai số là: {int(a) + int(b)}")
except ValueError:
    print("Giá trị bạn nhập không phải là một số nguyên! Hãy thử chạy lại")
```

#### Chương trình 2: Chương trình sử dụng vòng lặp cho đến khi người dùng nhập đúng số nguyên

```python
#Sử dụng While để yêu cầu người dùng nhập đến khi đúng
while True:
    try:
        a = int(input("Bạn hãy nhập giá trị đầu tiên: "))
        b = int(input("Bạn hãy nhập giá trị thứ hai:"))
        print(f"Kết quả là: {a+b}")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```

#### Chương trình 3: Giới hạn số lần nhập

```python
#Sử dụng While để yêu cầu người dùng nhập đến khi đúng, nếu quá 3 lần sẽ thoát chương trình
bo_dem = 0
while True:
    try:
        bo_dem += 1
        if bo_dem > 3:
            print("Bạn đã nhập quá 3 lần, chương trình đã thoát")
            break
        a = int(input("Bạn hãy nhập giá trị đầu tiên: "))
        b = int(input("Bạn hãy nhập giá trị thứ hai:"))
        print(f"Kết quả là: {a+b}, sau {bo_dem} lần thử")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```
