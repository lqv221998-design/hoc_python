<span style="color:orange"> 1. Tạo 2 biến a, b. Yêu cầu người dùng nhập từ bàn phím 2 số ngẫu nhiên. Sau đó in ra màn hình kết quả tổng của 2 số vừa nhập.</span>
#### Chương trình 1: Chương trình đơn giản, kèm xử lý ngoại lệ

```python
#Sử dụng hàm input, lệnh print, tính tổng, xử lý ngoại lệ:
a = input("Bạn hãy nhập giá trị đầu tiên: ") #Lỗi ValueError khi nhập chữ
b = input("Bạn hãy nhập giá trị thứ hai:")   #Lỗi ValueError khi nhập chữ
try:
    print(f"Tổng hai số là: {int(a) + int(b)}")
except ValueError:
    print("Giá trị bạn nhập không phải là một số nguyên! Hãy thử chạy lại")
```

#### Chương trình 2: Chương trình sử dụng vòng lặp cho đến khi người dùng nhập đúng số nguyên

```python
#Sử dụng While để yêu cầu người dùng nhập đến khi đúng
while True:
    try:
        a = int(input("Bạn hãy nhập giá trị đầu tiên: "))
        b = int(input("Bạn hãy nhập giá trị thứ hai:"))
        print(f"Kết quả là: {a+b}")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```

#### Chương trình 3: Giới hạn số lần nhập

```python
#Sử dụng While để yêu cầu người dùng nhập đến khi đúng, nếu quá 3 lần sẽ thoát chương trình
bo_dem = 0
while True:
    try:
        bo_dem += 1
        if bo_dem > 3:
            print("Bạn đã nhập quá 3 lần, chương trình đã thoát")
            break
        a = int(input("Bạn hãy nhập giá trị đầu tiên: "))
        b = int(input("Bạn hãy nhập giá trị thứ hai:"))
        print(f"Kết quả là: {a+b}, sau {bo_dem} lần thử")
        break
    except ValueError:
        print("Hãy nhập số nguyên")
```

<span style="color:orange">2. Tạo 2 biến a và b. Yêu cầu người dùng nhập vào từ bàn phím 2 số ngẫu nhiên. Kiểm tra a có chia hết cho b hay không? Nếu có, in ra màn hình dòng chữ  “a chia het cho b”, ngược lại, in ra dòng chữ “ a khong chia het cho b” (nhập b khác 0)
#### Cách 1 đơn giản, xử lý ngoại lệ 

```python
try:
    a = float(input("Nhập số a: "))
    b = float(input("Nhập số b: "))
    if a % b == 0:
        print('a chia hết cho b')
    else:
        print("a không chia hết cho b")
except ZeroDivisionError:
    print("Nhập b khác 0")
except ValueError:
    print("Bạn phải nhập một số thực")
```

#### cách 2: Vòng lặp yêu cầu người dùng nhập đến khi đúng

```python
while True:
    try:
        a = float(input("Nhập số a: "))
        b = float(input("Nhập số b: "))
        if a % b == 0:
            print('a chia hết cho b')
        else:
            print("a không chia hết cho b")
        break
    except ZeroDivisionError:
        print("Nhập b khác 0")
    except ValueError:
        print("Bạn phải nhập một số thực")
```

#### Cách 3: giới hạn số lần nhập 2 lần

```python
bo_dem = 0
while True:
    try:
        bo_dem += 1
        if bo_dem > 2:
            print("Đã hết số lần nhập")
            break
        a = float(input("Nhập số a: "))
        b = float(input("Nhập số b: "))
        if a % b == 0:
            print('a chia hết cho b')
        else:
            print("a không chia hết cho b")
        break
    except ZeroDivisionError:
        print("Nhập b khác 0")
    except ValueError:
        print("Bạn phải nhập một số thực")
```

<span style="color:orange">3. Tạo biến year yêu cầu người dùng nhập vào 1 số (giả sử người dùng luôn nhập số > 0 ). Hãy kiểm tra year và in ra màn hình năm đó có phải là năm nhuận hay không?
#### Cách 1 code chay

```python
# Logic: 
# Có 2 trường hợp cần phải xét:
# Năm không thế kỷ: năm đó chia hết cho 4
# Năm tròn thế kỷ (có 2 số 00 ở cuối ví dụ 2000): chia hết cho 400
try:
    year = input("Nhập năm")
    if year[2:]=='00':  #dùng slicing
        tinh = year % 400
        if tinh == 0:
            print(f"{year} là năm nhuận")
        else:
            print(f"{year} không phải là năm nhuận")
    elif int(year) == 0:
        print("bạn phải nhập năm khác 0")
    else:
        tinh = int(year) % 4
        if tinh == 0:
            print(f"{year} là năm nhuận")
        else:
            print(f"{year} không phải là năm nhuận")
except TypeError:
    print("Bạn phải nhập số lớn hơn 0")
except ValueError:
    print(f"{year} không phải là năm")
```

#### Cách 2 sử dụng thư viện

```python
import calendar as cl
def Tinh_năm_Nhuan(year):
    if cl.isleap(year) == True:
        return f"{year} là năm nhuận"
    else:
        return f"{year} không phải là năm nhuận"
try:
    year = int(input("Nhập năm"))
    print(Tinh_năm_Nhuan(year))
except:
    print("Nhập số nguyên")
```

6.  Viết chương trình kiểm tra xem tải trọng của dầm có vượt quá giới hạn 5000 kg không

```python
gioi_han = 5000
def kiem_tra_tai_trong(tai_trong_dam):
    if tai_trong_dam < gioi_han:
        return "Dầm đạt"
    else:
        return "Dầm vượt tải trọng"
try:
    tai_trong_dam = float(input("nhập tải trọng dầm"))
    print(kiem_tra_tai_trong(tai_trong_dam))
except ValueError:
    print("Nhập số thực")
```
